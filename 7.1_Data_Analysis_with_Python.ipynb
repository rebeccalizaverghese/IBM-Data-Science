## **Data Analysis with Python**
#### Importing Dataset


#### Data Acquisition
A data set is typically a file containing data stored in one of several formats (.csv, .json, .xlsx, etc) and can be stored in different locations (local machine, server, website, cloud storage, etc)

In order to analyse a dataset in a python notebook, we need to bring/ load the dataset into the notebook. We use the Pandas Library to do so.
The Pandas Library is a very popular and useful tool that enables us to read various datasets into a data frame.

```python
# Import pandas & numpy library
import pandas as pd
import numpy as np
```

```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
df = pd.read_csv(file_path, header=None)
# By specifying that 'header'=none, pandas automatically sets the column headers as a list of integers
```

```python
df.head(10) # provides the first 10 rows
```

```python
df.tail(5) # provides the last five rows
```

```python
# TO ADD HEADERS:
# create headers list
headers =["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]


df.columns = headers
df.columns
```

```python
df.head(3)
```

```python
# Replace '?' symbol with 'NaN' so that dropna() can remove missing values
df1 = df.replace('?', np.NaN)

# To drop missing values
df = df1.dropna(subset=['price'], axis=0) # (axis=0) means that the contents along the entire row will be dropped whereever the entity 'price' is found to be NaN
```

```python
df[['price']].head(10)
```

```python
# To find name of columns
print(df.columns)
```

```python
# to find type of columns
print(df.dtypes)
```

```python
# Saving Datasets
# Pandas enable us to save the dataset to csv. using the '.to_csv()' method
df.to_csv("automobile.csv", index=False)
```

```python
# Describe function - used to get statistical summary of each columns
df.describe()
```

```python
# To describe all columns
df.describe(include = 'all') # even shows object-typed attributes.
```

```python
# To find statistical summary for specific columns
df[['length', 'compression-ratio']].describe()
```

```python
# INFO - another method used to provide concise summary of dataframe - prints information about dataframe
df.info()
```
