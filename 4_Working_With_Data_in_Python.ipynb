##### Reading files with 'Open'

Reading text files involves extracting and processing the data stored within them. We use python's built-in 'open' function to get a file object and access the data within the file. The 'Open' function involves 2 primary parameters:
* File Path - consists of the file name and directory where the file is located
* Mode - specifies the purpose for opening the file - It includes 'r' for reading/ 'w' for writing and 'a' for appending.
Eg: open("/resources/data/Example1.txt"."r")

##### Writing Files with Open
Using the Open function, we can create a new text file and write data to it. In this case, the mode used will be the write 'w' mode

If we do not wish to create a new file but edit an existing file, we can use the append 'a' mode.

```python
eg2 = '/example2.txt'
with open(eg2, 'w') as writefile:
  writefile.write("This is line A\n")
  writefile.write("This is line B\n")
```

```python
# To read the file and see if it worked
with open(eg2, 'r') as testwritefile:
  print(testwritefile.read())
```

```python
# To write a list to a .txt file using loop
lines = ["This is line A\n", "This is line B\n", "This is line C"]

with open(eg2, 'w') as writefile:
  for line in lines:
    print(line)
    writefile.write(line)
```

```python
with open(eg2, 'r') as testwritefile:
  print(testwritefile.read())
```

```python
# It is important to note that setting the mode to w overwrites all the existing data in the file.
with open('/Example2.txt', 'w') as writefile:
    writefile.write("Overwrite\n")
with open('/Example2.txt', 'r') as testwritefile:
    print(testwritefile.read())
```

```python
# We can use the append 'a' mode to write files without losing exisiting data.
with open(eg2, 'a') as testwritefile:
    testwritefile.write("\nThis is line D\n")
    testwritefile.write("This is line E\n")
```

```python
with open(eg2, 'r') as testwritefile:
    print(testwritefile.read())
```

###### Additional Modes
It's fairly ineffecient to open the file in a or w and then reopening it in r to read any lines. Luckily we can access the file in the following modes:

* r+ : Reading and writing. Cannot truncate the file.
* w+ : Writing and reading. Truncates the file.
* a+ : Appending and Reading. Creates a new file, if none exists. You dont have to dwell on the specifics of each mode for this lab.

```python
with open(eg2, 'a+') as testwritefile:
    testwritefile.write("This is line E\n")
    print(testwritefile.read())
```

```python
with open('/Example2.txt', 'a+') as testwritefile:
    print("Initial Location: {}".format(testwritefile.tell()))

    data = testwritefile.read()
    if (not data):  #empty strings return false in python
            print('Read nothing')
    else:
            print(testwritefile.read())

    testwritefile.seek(0,0) # move 0 bytes from beginning.

    print("\nNew Location : {}".format(testwritefile.tell()))
    data = testwritefile.read()
    if (not data):
            print('Read nothing')
    else:
            print(data)

    print("Location after read: {}".format(testwritefile.tell()) )
```

```python
with open('/Example2.txt', 'r+') as testwritefile:
    testwritefile.seek(0,0) #write at beginning of file
    testwritefile.write("Line 1" + "\n")
    testwritefile.write("Line 2" + "\n")
    testwritefile.write("Line 3" + "\n")
    testwritefile.write("Line 4" + "\n")
    testwritefile.write("finished\n")
    testwritefile.seek(0,0)
    print(testwritefile.read())
```

##### Pandas:
Pandas is a popular open-source data manipulation and analysis library for the Python programming language. It provides a powerful and flexible set of tools for working with structured data. It is designed to handle data in various formats, such as tabular data, time-series data, etc.
It offers two primary data structures:
* Data Frame: two dimensional, tabular data structure comprising of rows and columns
* Series - One dimensional labeled array, comprising of a single column or row of data.

#### Working with NumPy
NumPy is a Python library used for working with arrays, linear algebra, fourier transform and matrices. NumPy stands for Numerical Python.
The array object in NumPy is called ndarray
You can import the NumPy library using the 'import' function.

```python
# A numpy array is fixed in size and each element is of the same type.
import numpy as np
a = np.array([0,1,2,3,4])
a
```

```python
# The elements can be called via their index
print('a[0]: ', a[0])
print('a[1]: ', a[1])
print('a[2]: ', a[2])
print('a[3]: ', a[3])
print('a[4]: ', a[4])
```

```python
# checking NumPy version - version string is stored under 'version' attribute
print(np.__version__)

# checking the type of the array
print(type(a))

# checking the type of values stored in numpy array
print("The data type of the array's elements is ",a.dtype)
```

```python
b = np.array([3.1, 11.02, 6.2, 213.2, 5.2])
print(b)
print(type(b))
print(b.dtype)
```
